EDA on Iris Dataset

```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Import dataset using pandas
df = pd.read_csv("Iris.csv",
                 delimiter = ",")
df.head()
```

Note : Here we kept csv dataset file in same or working directory

```python
# Checking the data types of all the columns
df.info()
```

```python
# Check the null/Missing Values in dataset
print(df.isnull().sum())
```

 0 represents here No missing values in dataset 

```python
# Check the descriptive statistics of numeric variables
df.describe().round(2)
```

Above table describe Sepal & Petal features show large range and variablity. 

Also excluding there different classes of species


```python
# Jointplot 2D of sepallength & sepalwidth including different species
sns.set_style("whitegrid")
plt.figure(figsize = (10,6))
sns.jointplot(data = df, x ='SepalLengthCm', y ='SepalWidthCm', hue = 'Species')
plt.show()
```

This plot visualizing the relationship between Sepal Length(cm) and Sepal Width(cm) for three different species:
   1. Setosa points are clearly separated from other two species based on sepal dimensions
   2. Virginica overlaps slightly with versicolor but have longer sepals 

```python
# Jointplot 2D of petallength & petalwidth including different species
sns.set_style("whitegrid")
plt.figure(figsize = (10,6))
sns.jointplot(data = df, x ='PetalLengthCm', y ='PetalWidthCm', hue = 'Species')
plt.show()
```

This joint plot easily classify based on Petal size

    1. Versicolor and virginica overlap at some region.
    2. Setosa is clearly separated from the other species.
    3. Virginica has the largest petals.

```python
# Counts based on different species
df['Species'].value_counts()
```

```python
# Violin plot based on Petal Length
sns.set_style("white")
sns.violinplot(data = df.drop(columns = 'Id'),x = 'Species', y = 'PetalLengthCm')
plt.title("Violinplot of Iris Dataset")
plt.ylabel("Centimeter")
plt.show()
```

This Violin plot based on Petal Length of Flowers

   1. Setosa is different from other two
   2. Versicolor and Virginica have more overlapping

```python
# Violin plot based on Petal Width
sns.set_style("white")
sns.violinplot(data = df.drop(columns = 'Id'),x = 'Species', y = 'PetalWidthCm')
plt.title("Violinplot of Iris Dataset")
plt.ylabel("Centimeter")
plt.show()
```

This is Violin plot based on Petal Width of three iris species

   1. Setosa < Versicolor < Virginica

```python
# Violin plot based on Sepal Length
sns.set_style("white")
sns.violinplot(data = df.drop(columns = 'Id'),x = 'SepalLengthCm', y = 'Species')
plt.title("Violinplot of Iris Dataset")
plt.xlabel("Centimeter")
plt.show()
```

This is Violin plot based on Sepal Length of three iris species

   1. The Wider Violin indicate where the data is denser.

```python
# Violin plot based on Sepal Width
sns.set_style("white")
sns.violinplot(data = df.drop(columns = 'Id'),x = 'SepalWidthCm', y = 'Species')
plt.title("Violinplot of Iris Dataset")
plt.ylabel("Centimeter")
plt.show()
```

This is Violin plot based on Sepal Width :

   1. For all Violin plot, Median values is shown by white dots in the violin.
   2. Setosa has the highest median sepal width.

```python
# Visualize multiple variables in a dataset
sns.pairplot(data = df.drop(columns = 'Id'), hue = 'Species')
```

Above plot is seaborn Pairplot, which shows Bivariate relation.

   1. Setosa species is separated from the other two across all feature combinations.
